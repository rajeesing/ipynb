```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.describe()
```

```python
plib.scatter(iris["sepal_length"], iris["sepal_width"])
plib.xlabel("Sepal Length")
plib.ylabel("Sepal Width")
plib.title("Sepal Length vs Sepal Width")
```

```python
plib.boxplot(iris["petal_length"])
plib.title("Sepal Length Distribution")
```

```python
plib.hist(iris["petal_width"], bins=20)
plib.title("Sepal Length Distribution")
```

```python
spicies_count = iris["species"].value_counts()
plib.pie(spicies_count, labels=spicies_count.index, autopct='%1.1f%%')
plib.title("Sepcies Distribution")
```

```python
sns.set_theme(style="ticks")
sns.pairplot(iris, hue="species")
```

```python
sns.set_theme(style="whitegrid")
sns.set_palette(sns.color_palette("pastel"))
fig, ax = plib.subplots(ncols=2, nrows=2, figsize=(15, 6))
sns.boxplot(x="sepal_length", hue="species", data=iris, ax=ax[0, 0])
sns.violinplot(y="sepal_length", x="species", data=iris, ax=ax[0, 1])
sns.histplot(y="sepal_length", x="species", data=iris, ax=ax[1, 1])
sns.barplot(y="sepal_length", x="species", data=iris, ax=ax[1, 0])
```

```python
#linear model - regression
sns.set_theme(style="whitegrid")
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue="species")

```

```python
sns.heatmap(iris.corr(), annot=True)
plib.show()
```
